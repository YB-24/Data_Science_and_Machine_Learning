```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = x * 2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
#large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("power of 4")
#small axes
axes2 = fig.add_axes([0.2,0.3,0.25,0.25])
axes2.plot(x,y)
axes2.set_xlim(1,2)
axes2.set_ylim(0,25)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title("Zoomed In")
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(10,3),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('myfirstfigure',bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,10),dpi=200)

axes[0][0].plot(x,y)
axes[0][0].set_title('hiro')

axes[1][1].plot(a,b)
axes[1][1].set_xlabel('hello')

fig.suptitle('Figure Level Title',fontsize=16)

#plt.tight_layout()-- this will automatically create space between graphs so that they dont overlap this should be in the last line of cell

#if we want to manually adjust the space between the graphs then we have to use another command which is also written at the last of the cell
fig.subplots_adjust(wspace=0.9,hspace=0.3)  #the parameters are the fraction of average width and average height of graphs axes

fig.savefig('new_subplots.png',bbox_inches='tight')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5))   #see doc for loc argument, we can pass the tuple for manual positioning with reference to respective axes fraction
```

```python
fig = plt.figure(figsize=(3,3))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',label='x v x')  #either give color name or give its hex code

ax.plot(x,x+1,color='purple',label='x v x+1')

ax.legend()
```

```python
fig = plt.figure(figsize=(3,3))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=1.5,ls='--',marker='s',markersize=5)  #lw and linewidth are same;for built-in linestyle and markers see docs

ax.plot(x,x+1,color='purple',linewidth=2,linestyle='-.')
```
