```python
X = pd.get_dummies(df.drop('species',axis=1),dtype=np.number,drop_first=True)
```

```python
y = df['species']   #no need of scaling as we are only checking conditions on node
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,base_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['feature importance']).sort_values('feature importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10,8),dpi=200)
plot_tree(model);
```

```python
plt.figure(figsize=(10,8),dpi=400)
plot_tree(model,feature_names=X.columns,filled=True);
```

```python
#to ease our task create a plot function
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(10,8),dpi=400)
    plot_tree(model,feature_names=X.columns,filled=True);
```

```python
report_model(model)
```

#### Playing with the hyperparameters

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train,y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')  #another criterion instead of gini
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
