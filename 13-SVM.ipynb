#### SVM for Regression

```python
df = pd.read_csv('cement_slump.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.heatmap(df.corr(),annot=True)
```

```python
X = df.drop('Compressive Strength (28-day)(Mpa)',axis=1)
y = df['Compressive Strength (28-day)(Mpa)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler  #strictly needed for this
```

```python
scaler= StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVR,LinearSVR     #LinearSVR when kernel='linear'  
```

```python
base_model = SVR()
```

```python
base_model.fit(X_train,y_train)
```

```python
base_pred = base_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,base_pred)
```

```python
np.sqrt(mean_squared_error(y_test,base_pred))
```

```python
#perform grid search to find a better model
```

```python
param_grid = {'C':[0.001,0.01,0.1,0.5,1],'kernel':['linear','rbf','poly'],
             'gamma':['scale','auto'],'degree':[2,3,4],
             'epsilon':[0,0.01,0.1,0.5,1,2]}   #epsilon means the error we allow to occour
```

```python
grid_svr = GridSearchCV(base_model,param_grid)
```

```python
grid_svr.fit(X_train,y_train)
```

```python
grid_svr.best_params_
```

```python
grid_pred = grid_svr.predict(X_test)
```

```python
mean_absolute_error(y_test,grid_pred)
```

```python
mean_squared_error(y_test,grid_pred)**0.5
```
